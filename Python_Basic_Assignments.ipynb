```python
#Q4 discuss the different types of operator in python and provide examples of how they used

Ans: Python has several types of operators that are used to perform operations on variables and values.

1. Arithmetic Operators

Perform basic mathematical operations.

Example:
    
a= 10
b = 3
print(a + b)  # Addition, Output: 13
print(a - b)  # Subtraction, Output: 7
print(a * b)  # Multiplication, Output: 30
print(a / b)  # Division, Output: 3.3
print(a % b)  # Modulus, Output: 1
print(a ** b) # Exponentiation

2. Comparison Operators

Compare two values and return a boolean result.

a = 10
b = 5
print(a == b)  # Equal, Output: False
print(a != b)  # Not equal, Output: True
print(a > b)   # Greater than, Output: True
print(a < b)   # Less than, Output: False
print(a >= b)  # Greater than or equal to, Output: True
print(a <= b)  # Less than or equal to, Output : false

3. Logical Operators

Combine conditional statements.

a = True
b = False
print(a and b)  # Logical AND, Output: False
print(a or b)   # Logical OR, Output: True
print(not a)    # Logical NOT,

4. Assignment Operators

Assign values to variables with optional arithmetic.

a = 10
a += 5  # Output: 15
a -= 2  # Output: 13
a *= 3  # Output: 39
a /= 3


5. Bitwise Operators

 Perform bit-level operations on binary numbers.
    
a = 10 
b = 4  
print(a & b)  # Bitwise AND, Output: 0
print(a | b)  # Bitwise OR, Output: 14
print(a ^ b)  # Bitwise XOR, Output: 14
print(~a)     # Bitwise NOT, Output: -11
print(a << 1) # Bitwise left shift, Output: 20
print(a >> 1) # Bitwise right shift, Output: 5
    
6. Identity Operators

 Check if two variables reference the same object.
    
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)   # Output: True
print(a is c)   # Output: False
print(a is not c)  # Output: True

7. Membership Operators

Test if a sequence contains a specified value.

a = [1, 2, 3]
print(2 in a)     # Output: True
print(5 in a)     # Output: False
print(6 not in a) # Output: True


```

```python
#Q5 Explain the concept of typecasting in python with examples ?

Ans: Typecasting resfers to converting one data type to another. Built in function such as int(), float(), str(), and list().
used for conversion 

Examples 

Integer to String 

num =6;
Str_num = str(num)

print(Str_num)  #output : '6'
print(type(str_num))  # output: <String>

String to Integer:
    
str_num = "14"
num = int(str_num)
print(num)              # Output: 14
print(type(num))        # Output: <int'>

List to Tuple:
    
lst = [1, 2, 3]
tpl = tuple(lst)
print(tpl)              # Output: (1, 2, 3)
print(type(tpl))        # Output: <class 'tuple'>

```
